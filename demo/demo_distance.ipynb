```python
import sys
sys.path.append("../")
import numpy as np
import matplotlib.pyplot as plt
from pyACC.cosmology import Distances
```

```python
H0 = 70
OmegaM = 0.3
OmegaL = 0.7
OmegaR = 0.0

d = Distances(H0, OmegaM, OmegaL, OmegaR)

zvals = np.linspace(0, 100, 1000)
plt.plot(zvals, [d.comoving_distance(z) for z in zvals], 'r.')
plt.xlabel("z")
plt.ylabel("Comoving distance [Mpc]")
plt.show()
```

```python

plt.plot(zvals, [d.angolar_distance(z) for z in zvals], 'r.')
plt.xlabel("z")
plt.ylabel("Angolar distance (Mpc)")
plt.show()
```

```python
plt.plot(zvals, [d.luminosity_distance(z) for z in zvals], 'r.')
plt.xlabel("z")
plt.ylabel("Luminosity distance (Mpc)")
plt.show()
```

```python
for z in np.linspace(0, 100, 1000):
    plt.plot(z, d.angolar_distance(z), 'r.')
plt.xlabel("z")
plt.ylabel("Angolar distance (Mpc)")
plt.show()
```

```python
for z in np.linspace(0, 100, 1000):
    plt.plot(z, d.luminosity_distance(z), 'r.')
plt.xlabel("z")
plt.ylabel("Luminosity distance (Mpc)")
plt.show()
```
