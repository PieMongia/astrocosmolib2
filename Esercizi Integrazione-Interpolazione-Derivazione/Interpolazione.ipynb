```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fx = lambda x: np.sin(x)
#Analoga a definire una funzione, a sx dei due punti ci sono gli argomenti mentre a destra cosa fa la funzione
```

```python
x_points=np.linspace(0, 2*np.pi, 100)
y_points=fx(x_points)

plt.plot(x_points, y_points)
plt.show()
```

```python
def linear_interpolation(x, x_nodes, y_nodes):
    index = np.digitize(x, x_nodes)-1
    #compute the interpolated valude
    return y_nodes[index] + (y_nodes[index+1]-y_nodes[index])/(x_nodes[index+1]-x_nodes[index])*(x-x_nodes[index])
```

```python
x_nodes = np.linspace(0,2*np.pi,20)
y_nodes = fx(x_nodes)

x_interp= np.linspace( 0, 2*np.pi*0.999, 200)
y_interp = linear_interpolation(x_interp, x_nodes, y_nodes)
print(y_interp.shape)
plt.plot(x_points,y_points, color='b')
plt.plot(x_interp,y_interp,'r')
plt.plot(x_nodes, y_nodes, 'ob')
#plt.show()
```

```python
y_true=fx(x_interp)
diff=np.abs(y_interp-y_true)
```

```python
plt.plot(x_interp, diff, lw=3)
plt.xlabel("x")
plt.ylabel("Error")

for i in range(len(x_nodes)):
  plt.axvline(x=x_nodes[i], color='r', ls='--')

plt.plot(x_interp, np.abs(np.sin(x_interp)*(x_nodes[1]-x_nodes[0])**2 / 8))
```
