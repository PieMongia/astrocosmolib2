```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.integrate import trapezoid, simpson, quad
```

```python
def function(x):
  return x**2 * np.sin(x) * np.exp(-x)
```

```python
xx = np.logspace(-4, 3, 1000)
fx=function(xx)
```

```python
plt.semilogx(xx,fx)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
Integral_quad = quad(function, 1.e-4, 1000, epsabs=1.e-10 ,epsrel=1.e-10, limit=10000)
```

```python
Integral_quad
```

```python
xx = np.logspace(-4, 3, 10000000)
fx=function(xx)
```

```python
Integral_trapz= trapezoid(fx, xx)
Integral_trapz, Integral_quad[0]
```

```python
plt.axhline(y=Integral_quad[0], color='r', linestyle='--')
list_n=np.logspace(2,4, 20)
list_t=[]
list_s=[]
for n in list_n:
  xx=np.linspace(0,100,int(n))
  fx=function(xx)
  Integral_trapz=trapezoid(fx,xx)
  Integral_simps=simpson(fx,x=xx)
  list_t.append(Integral_trapz)
  list_s.append(Integral_simps)

plt.loglog(list_n, np.abs(list_t), label="")
plt.loglog(list_n, np.abs(list_s), label="Simpson")
plt.xlabel("n")
plt.ylabel("Integral")
plt.show()
```
